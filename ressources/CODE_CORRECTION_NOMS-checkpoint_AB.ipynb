### supression des accents

```python
import unicodedata
def remove_accents(input_str):
    # Normaliser la chaîne de caractères en forme décomposée (NFD)
    normalized_str = unicodedata.normalize('NFD', input_str)
    # Filtrer les caractères accentués en gardant seulement les caractères ASCII
    return ''.join(c for c in normalized_str if unicodedata.category(c) != 'Mn')

# Appliquer la fonction de suppression des accents à la colonne existante
df['SN HARMONISE'] = df['SN HARMONISE'].apply(remove_accents)
df.head(10)
```

### calcul de la matrice de correlation
##### 1- Normalisation des Noms
Avant de comparer les noms, il est important de les normaliser. La normalisation peut inclure des étapes comme trier les mots dans les noms pour gérer les permutations (par exemple, "JOHN SMITH" et "SMITH JOHN" deviennent identiques après tri). Cela aide à réduire les variations dues à l'ordre des mots et à améliorer la précision des comparaisons.

##### 2-Transformation en Représentation Numérique
Les noms normalisés sont ensuite transformés en une représentation numérique que les ordinateurs peuvent manipuler. Cette étape utilise un modèle appelé TF-IDF (Term Frequency-Inverse Document Frequency).TF-IDF évalue l'importance de chaque caractère ou groupe de caractères dans un nom par rapport à tous les noms dans la base de données.

-Term Frequency (TF) : Mesure combien de fois un caractère ou un groupe de caractères apparaît dans un nom.

-Inverse Document Frequency (IDF) :Mesure combien un caractère ou un groupe de caractères est rare dans l'ensemble des noms. 

En combinant ces deux mesures, TF-IDF crée un vecteur pour chaque nom, qui représente la présence et l'importance des différents caractères ou groupes de caractères.

##### 3- Calcul de la Similarité Cosinus

La similarité cosinus est utilisée pour mesurer la similarité entre les vecteurs de noms. En termes simples, elle évalue à quel point deux vecteurs (représentations numériques des noms) sont similaires en termes de direction. La similarité cosinus mesure l'angle entre deux vecteurs. Si l'angle est petit (proche de 0 degré), les vecteurs sont très similaires et donc les noms sont considérés comme similaires. Si l'angle est grand (près de 90 degrés), les vecteurs sont très différents, ce qui signifie que les noms ne sont pas similaires.

##### 4- Création de la Matrice de Similarité
Une fois que la similarité cosinus est calculée entre chaque paire de noms, ces valeurs sont organisées dans une matrice de similarité. Cette matrice montre la similarité entre tous les noms par paire.

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Fonction pour normaliser les noms (tri des mots)
def normalize_name(name):
    words = name.split()
    words.sort()  # Trie les mots pour gérer les permutations
    return ' '.join(words)

# Application de la normalisation
df['Normalized Name'] = df['SN HARMONISE'].apply(normalize_name)

# Représentation vectorielle avec TF-IDF
vectorizer = TfidfVectorizer(analyzer='char', ngram_range=(2, 3))
tfidf_matrix = vectorizer.fit_transform(df['Normalized Name'])

# Calcul de la similarité cosinus
cosine_similarities = cosine_similarity(tfidf_matrix)

# Matrice de similarité
cosine_df = pd.DataFrame(cosine_similarities, index=df['SN HARMONISE'], columns=df['SN HARMONISE'])
cosine_df
```

### créer des groupes basée sur la similarité

```python
##ATTRIBUER LE PREMIER NOM RENCONTRE

import numpy as np
# Seuil de similarité
threshold = 0.80

# Étape 1: Créer des groupes basés sur la similarité
group_ids = np.zeros(df.shape[0])

for i in range(len(cosine_similarities)):
    for j in range(i + 1, len(cosine_similarities)):
        if cosine_similarities[i, j] >= threshold:
            group_ids[j] = group_ids[i] if group_ids[i] != 0 else i + 1
            group_ids[i] = group_ids[i] if group_ids[i] != 0 else i + 1

# Étape 2: Si un nom n'a pas été groupé, lui attribuer un identifiant unique
unique_id = max(group_ids) + 1
for i in range(len(group_ids)):
    if group_ids[i] == 0:
        group_ids[i] = unique_id
        unique_id += 1

# Étape 3: Ajouter la nouvelle colonne 'nom_groupé' avec le nom commun par groupe
df['nom_final'] = df.groupby(group_ids)['SN HARMONISE'].transform('first')

df.head(10)
```
