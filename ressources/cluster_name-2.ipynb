```python
from fuzzywuzzy import process
import pandas as pd

df = pd.read_excel("Test_doublon.xlsx")
df = df.head(30)
df
```

```python
def elements_similaire(row, df, column_name="SN HARMONISE", seuil=80):
    nom = row[column_name]
    liste_similaire = []
    autres = df.loc[df[column_name] != nom, column_name].tolist()
    resultat = process.extractBests(nom, autres)
    for nom_similaire, score in resultat:
        if score >= seuil:
            liste_similaire.append(nom_similaire)
    if liste_similaire:
        return max(liste_similaire, key=len)
    else:
        return None

df["NOM_SIMILAIRE"] = df.apply(lambda row: elements_similaire(row, df, column_name="SN HARMONISE"), axis=1)
df
```
