```python
# Identifier les colonnes numériques (toutes sauf PAYS et ANNEE)
num_cols = [col for col in df.columns if col not in ["PAYS", "ANNEE"]]
print(num_cols)
```

```python
# S'assurer que les colonnes sont bien numériques
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors="coerce")
# Remplacer les NaN par la moyenne du pays + corriger le FutureWarning avec copy=False
df[num_cols] = df.groupby("PAYS")[num_cols].transform(
    lambda col: col.fillna(col.mean()).infer_objects(copy=False))
```

```python
# Application de l'ACP pour réduction de dimension
pca1 = PCA()
X_pca = pca1.fit_transform(df_macro_scaled)
X_pca_macro_df = pd.DataFrame(X_pca, columns=[f"PC{i+1}" for i in range(X_pca.shape[1])])
```

```python
# Détermination du nombre de composantes pour atteindre 80% de variance expliquée
explained_variance1 = np.cumsum(pca1.explained_variance_ratio_)
n_components1 = np.argmax(explained_variance1 >= 0.80) + 1
print(f"Nombre de composantes à retenir : {n_components1}")
```

```python
# Détermination du nombre de composantes pour atteindre 80% de variance expliquée
explained_variance2 = np.cumsum(pca2.explained_variance_ratio_)
n_components2 = np.argmax(explained_variance2 >= 0.80) + 1
print(f"Nombre de composantes à retenir : {n_components2}")
```

```python
# Détermination du nombre de composantes pour atteindre 80% de variance expliquée
explained_variance3 = np.cumsum(pca3.explained_variance_ratio_)
n_components3 = np.argmax(explained_variance3 >= 0.80) + 1
print(f"Nombre de composantes à retenir : {n_components3}")
```

```python
# Détermination du nombre de composantes pour atteindre 80% de variance expliquée
explained_variance4 = np.cumsum(pca4.explained_variance_ratio_)
n_components4 = np.argmax(explained_variance4 >= 0.80) + 1
print(f"Nombre de composantes à retenir : {n_components4}")
```

```python
# ACP FINALE
pca_final1 = PCA(n_components=n_components1)
X_pca_final1 = pca_final1.fit_transform(df_macro_scaled)
X_pca_final_df1 = pd.DataFrame(X_pca_final1, columns=[f"PC{i+1}" for i in range(n_components1)])
X_pca_final_df1
```

```python
# Sélection automatique des colonnes PC*
colonnes_pc = [col for col in X_pca_final_df1.columns if col.startswith("PC")]
# Calcul du Score Global par moyenne simple
X_pca_final_df1["PC1_Global"] = X_pca_final_df1[colonnes_pc].mean(axis=1)
X_pca_final_df1
```

```python
df_resultat2 = pd.concat([
    df_Système_financier[["PAYS", "ANNEE"]].reset_index(drop=True),
    X_pca_final_df2[["PC2_Global"]].reset_index(drop=True)
], axis=1)
df_resultat2["Dimension2"] = "Système financier"
df_resultat2
```

```python
# ACP FINALE
pca_final3 = PCA(n_components=n_components3)
X_pca_final = pca_final3.fit_transform(df_inclusion_scaled)
X_pca_final_df3 = pd.DataFrame(X_pca_final, columns=[f"PC{i+1}" for i in range(n_components3)])
X_pca_final_df3
```

```python

# RENOMMAGE DES COLONNES DE SCORE POUR UNIFORMISER
df_resultat1 = df_resultat1.rename(columns={"PC1_Global": "Macroéconomie"})
df_resultat2 = df_resultat2.rename(columns={"PC2_Global": "Système financier"})
df_resultat3 = df_resultat3.rename(columns={"PC3_Global": "Inclusion financière et numérique"})
df_resultat4 = df_resultat4.rename(columns={"PC4_Global": "Démographie et développement"})

# FUSION DES BASES PAR DIMENSION
df_merge = pd.merge(df_resultat1, df_resultat2, on=["PAYS", "ANNEE"], how="outer")
df_merge = pd.merge(df_merge, df_resultat3, on=["PAYS", "ANNEE"], how="outer")
df_merge = pd.merge(df_merge, df_resultat4, on=["PAYS", "ANNEE"], how="outer")

# TRANSFORMATION EN FORMAT LONG
df_final = pd.melt(df_merge, id_vars=["PAYS", "ANNEE"], 
                   value_vars=["Macroéconomie", "Système financier", "Inclusion financière et numérique", "Démographie et développement"],
                   var_name="Dimension", value_name="PC_Score")
df_final 
```

```python
# COMPTAGE DES VARIABLES PAR DIMENSION
compte_dimensions = {nom: len(colonnes) for nom, colonnes in DIMENSIONS.items()}

# CONVERSION EN DATAFRAME POUR AFFICHAGE
df_compte = pd.DataFrame(list(compte_dimensions.items()), columns=["Dimension", "Nombre_de_variables"])
df_compte
```

```python
# PONDÉRATIONS
reponse = input("Voulez-vous entrer vos propres pondérations pour les dimensions ? (oui/non) ").strip().lower()
if reponse == "oui":
    poids_macro = float(input("Pondération pour Macroéconomie = "))
    poids_financier = float(input("Pondération pour Système financier = "))
    poids_Inclusion = float(input("Pondération pour Inclusion financière et numérique = "))
    poids_demo = float(input("Pondération pour Démographie et développement = "))
else:
    #determination automatique 
    poids_macro = poids_financier = poids_Inclusion = poids_demo = 1*100/len(df_compte.Dimension.unique())
```

```python
# CALCUL DU SCORE PONDÉRÉ
df_final ['Pondération'] = df_final ['Dimension'].map({
    'Macroéconomie': poids_macro,
    "Système financier": poids_financier,
    'Inclusion financière et numérique': poids_Inclusion,
    'Démographie et développement': poids_demo
})
df_final ['Score_dimension'] = df_final ['PC_Score'] * (df_final ['Pondération']/ 100)
```

```python
df_final 
```

```python
min_score = df_final['Score_dimension'].min()
max_score = df_final['Score_dimension'].max()
df_final['score_normalise'] = 0+ ((df_final['Score_dimension'] - min_score) / (max_score - min_score)) * 100
```

```python
df_final.head(50)
```

```python
# pour fais la somme des dimensions normaliser 
df_final = df_final.groupby(["PAYS", "ANNEE"])["score_normalise"].sum().reset_index()
df_final
```

```python
# Division par 4 dimensions
df_final["Score_Global"] = df_final["score_normalise"] / 4
# AFFICHAGE DU RÉSULTAT
df_final.head(20)
```

```python
def attribuer_notation(score):
    if 75 <= score <= 100:
        return "AAA", "Excellente qualité, risque minimal"
    elif 50 <= score < 75:
        return "AA-", "Haute qualité, risque faible"
    elif 25 <= score < 50:
        return "B+", "Hautement spéculatif, risque élevé"
    elif 0 <= score < 25:
        return "D", "Défaut avéré"
    else:
        return None, "Score invalide"
```

```python
df_final[['Notation', 'Signification']] = df_final['Score_Global'].apply(
    lambda x: pd.Series(attribuer_notation(x))
)
```
